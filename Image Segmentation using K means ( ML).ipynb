```python
from PIL import Image
import matplotlib.pyplot as plt

# Path to the image on your laptop
image_path = 'C:/Users/gyanada/Downloads/Image.jpeg'
try:
    # Open the image using PIL
    image = Image.open(image_path)

    # Display the image
    plt.imshow(image)
    plt.axis('off')  
    plt.show()

except FileNotFoundError:
    print(f"Error: The file '{image_path}' was not found. Please check the file path.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
image_array = np.array(image)
# Reshape the image into a 2D array of pixels and 3 color values (RGB)
pixel_vals = image_array.reshape((-1, 3))
# Convert pixel values to float32 type
pixel_vals = np.float32(pixel_vals)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85)
k = 3
retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
segmented_image = segmented_data.reshape(image.size[1], image.size[0], 3)
plt.imshow(segmented_image)
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85)
k = 15
retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
segmented_image = segmented_data.reshape(image.size[1], image.size[0], 3)
plt.imshow(segmented_image)
plt.axis('off')
plt.show()
```
